```python
import time 
import requests
import cv2
import operator
import numpy as np

# Import library to display results
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
%matplotlib inline 
# Display images within Jupyter
```

```python
# Variables
_trainStatusUrl = 'https://api.cognitive.azure.cn/face/v1.0/persongroups/{personGroupId}/training'
_trainGroupUrl = 'https://api.cognitive.azure.cn/face/v1.0/persongroups/{personGroupId}/train'
_addFaceUrl = 'https://api.cognitive.azure.cn/face/v1.0/persongroups/{personGroupId}/persons/{personId}/persistedFaces'
_createGroupUrl = 'https://api.cognitive.azure.cn/face/v1.0/persongroups/{personGroupId}'
_createGroupPersonUrl = 'https://api.cognitive.azure.cn/face/v1.0/persongroups/{personGroupId}/persons'
_verifyUrl = 'https://api.cognitive.azure.cn/face/v1.0/verify'
_detectUrl = 'https://api.cognitive.azure.cn/face/v1.0/detect?returnFaceId=true'
_key = None  #Here you have to paste your primary key
_key = "f4af9871a9ed442fa6ab1322d38f16a4" #Here you have to paste your primary key
_maxNumRetries = 10
```

# Helper functions

```python
def processRequest(url, json, data, headers, params, method):

    """
    Helper function to process the request to Project Oxford

    Parameters:
    json: Used when processing images from its URL. See API Documentation
    data: Used when processing image read from disk. See API Documentation
    headers: Used to pass the key information and the data type request
    """

    retries = 0
    result = None

    while True:
        response = requests.request( method, url, json = json, data = data, headers = headers, params = params )

        if response.status_code == 429:
            if retries <= _maxNumRetries: 
                time.sleep(1) 
                retries += 1
                continue
            else: 
                print( 'Error: failed after retrying!' )
                break
        elif response.status_code == 202 or response.status_code == 200:
            if(response.content is not None and response.json() is not None):
                result = response.json()
                print("Message: %s" % ( response.json()))
            else:
                result = response
        else:
            print( "Error code: %d" % ( response.status_code ) )
            result = response
        break
    
    print( "Status code: %d" % ( response.status_code ) )
    return result
```

```python
def createPersonGroup(groupName):
    url = _createGroupUrl.replace("{personGroupId}", groupName)
    params = None
    json = dict()
    json['name'] = groupName
    data = None
    headers = dict()
    headers['Ocp-Apim-Subscription-Key'] = _key
    headers['Content-Type'] = 'application/json'
    return processRequest(url, json, data, headers, params, 'put')
    
def createPersonGroupPerson(groupName, personName):
    url = _createGroupPersonUrl.replace("{personGroupId}", groupName)
    params = None
    json = dict()
    json['name'] = personName
    data = None
    headers = dict()
    headers['Ocp-Apim-Subscription-Key'] = _key
    headers['Content-Type'] = 'application/json'
    return processRequest(url, json, data, headers, params, 'post')
    
def trainPersonGroup(groupName):
    url = _trainGroupUrl.replace("{personGroupId}", groupName)
    params = None
    json = None
    data = None
    headers = dict()
    headers['Ocp-Apim-Subscription-Key'] = _key
    headers['Content-Type'] = 'application/json'
    return processRequest(url, json, data, headers, params, 'post')

def getTrainStatus(groupName):
    url = _trainStatusUrl.replace("{personGroupId}", groupName)
    params = None
    json = None
    data = None
    headers = dict()
    headers['Ocp-Apim-Subscription-Key'] = _key
    headers['Content-Type'] = 'application/json'
    return processRequest(url, json, data, headers, params, 'get')
    
def addFaces(groupName, personName, photoPaths):
    url = _addFaceUrl.replace("{personGroupId}", groupName)
    url = url.replace("{personId}", personName)
    for path in photoPaths:
        pathToFileInDisk = path
        with open(pathToFileInDisk, 'rb') as f:
            data = f.read()
        
        params = None
        json = None
        headers = dict()
        headers['Ocp-Apim-Subscription-Key'] = _key
        headers['Content-Type'] = 'application/octet-stream'
        
        json = None
        result = processRequest(url, json, data, headers, params, 'post')
```

```python
def verifyPersonGroup(groupName, personName, targetFaceId):
    url = _verifyUrl
    params = dict()
    params['faceId'] = targetFaceId
    params['personId'] = personName
    params['personGroupId'] = groupName
    
    json = None
    data = None
    headers = dict()
    headers['Ocp-Apim-Subscription-Key'] = _key
    headers['Content-Type'] = 'application/json'
    return processRequest(url, json, data, headers, params, 'post')
```

```python
def showResultOnImage( result, img ):
    
    """Display the obtained results onto the input image"""
    img = img[:, :, (0, 1, 2)]
    fig, ax = plt.subplots(figsize=(12, 12))
    ax.imshow(img, aspect='equal')

    for face in result: 
        rect = face['faceRectangle']
        tl = (rect['left'], rect['top'])
        tr = (rect['left'] + rect['width'], rect['top'])
        bl = (rect['left'], rect['top'] + rect['height'])
        br = (rect['left'] + rect['width'], rect['top'] + rect['height'])
        text = face['faceId']
        x = [tl[0], tr[0], tr[0], br[0], br[0], bl[0], bl[0], tl[0]]
        y = [tl[1], tr[1], tr[1], br[1], br[1], bl[1], bl[1], tl[1]]
        line = Line2D(x, y, linewidth=3.5, color='red')
        ax.add_line(line)
        ax.text(tl[0], tl[1] - 2, '{:s}'.format(text),
        bbox=dict(facecolor='blue', alpha=0.5),
        fontsize=14, color='white')

    plt.axis('off')
    plt.tight_layout()
    plt.draw()
    plt.show()
```

# Analysis of an image stored on disk

```python
# Load raw image file into memory
pathToFileInDisk = r'/Users/h177954/Desktop/Personal/photo.jpg'
with open(pathToFileInDisk, 'rb') as f:
    data = f.read()

# Computer Vision parameters
params = None

headers = dict()
headers['Ocp-Apim-Subscription-Key'] = _key
headers['Content-Type'] = 'application/octet-stream'

json = None

result = processRequest(_detectUrl, json, data, headers, params, 'post')

faceId = result[0]["faceId"]

# Load the original image, fetched from the URL
if result is not None and len(result) > 0:
    data8uint = np.fromstring(data, np.uint8)  # Convert string to an unsigned int array
    img = cv2.cvtColor(cv2.imdecode(data8uint, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB)
    showResultOnImage(result, img)
```
